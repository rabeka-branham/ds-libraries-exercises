### Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import env
def get_db_url(username, host_name, password, database_name):
    return f'mysql+pymysql://{username}:{password}@{host_name}/{database_name}'
url = get_db_url(env.username, env.host, env.password, 'chipotle')
```

```python
df5 = pd.read_sql('select * from orders', url)
df5['item_price'] = df5.item_price.str.replace('$','').astype('float')
df5
```

```python
df6 = df5.groupby('item_name')[['quantity','item_price']].sum().sort_values('quantity',ascending=False).head(4)
df6 = df6.reset_index()
df6
```

```python
sns.barplot(data=df6,y='item_name',x='item_price');
```
