```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

### 2. Create and label 4 separate charts with the following equations (choose a range for x that makes sense):



### You can use functions from the math module to help implement some of the equations above.

```python
x1 = list(range(0,11))
y1 = [math.sqrt(n) for n in x1]
plt.plot(x1,y1)
plt.title('Square Root of X')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x2 = list(range(0,11))
y2 = [n**3 for n in x2]
plt.plot(x2,y2)
plt.title('x-cubed')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x3 = list(range(0,11))
y3 = [2**n for n in x3]
plt.plot(x3,y3)
plt.title('2 Raised to the x-Power')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x4 = list(range(0,11))
y4 = [1/(n+1) for n in x4]
plt.plot(x4,y4)
plt.title('1 divided by x + 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(x1,y1)
ax[0,1].plot(x2,y2)
ax[1,0].plot(x3,y3)
ax[1,1].plot(x4,y4)

fig.supxlabel('x-axis') 
fig.supylabel('y-axis')
fig.suptitle('Subplots')
fig.tight_layout()

plt.show()
```
